## f. O programa 3 deverá ser dividido em 3 células para tratar stopwords. Na primeira célula efetue a importação do pacote de stopwords do nltk.

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
```

## g. Na  segunda  célula  do  programa  3  defina  uma  função  para calcular  o  percentual  de  palavras  que  não  são stopwords no text3.

```python
def mfc_stopwords(text):
  mfc_stopwords = nltk.corpus.stopwords.words('english')
  mfc_content = [w for w in text if w.lower() not in mfc_stopwords]
  return (len(mfc_content) / len(text)) * 100
```

## h. Na terceira célula do programa 3 faça a chamada da função da célula definida na atividade anterior (g).

```python
mfc_stopwords(text3)
```
