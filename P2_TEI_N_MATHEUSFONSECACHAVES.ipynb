```python
from urllib import request
mfc_url = "http://www.gutenberg.org/files/5200/5200-0.txt"
mfc_response = request.urlopen(mfc_url)
mfc_raw = mfc_response.read().decode('utf8')
mfc_raw[0:1000]
```

**C) Na terceira célula deverá constar o código para tokenização do arquivo bruto e armazenamento em uma
variável do tipo Text do NLTK e a impressão em tela dos 98 primeiros tokens.** 


```python
import nltk
nltk.download('punkt')
mfc_tokens = word_tokenize(mfc_raw)
mfc_text = nltk.Text(mfc_tokens)
mfc_text [:98]
```

**D) Na quarta célula deverá ser realizada a normalização dos tokens e a construção do vocabulário, conforme
pipeline do PLN trabalhado em sala de aula.**

```python
mfc_words = [w.lower() for w in mfc_text]
mfc_vocab = sorted(set(mfc_words))
```

**B) Na segunda célula gerar uma lista impressa das etiquetas (tags) mais comuns do corpus mac_morpho**.

```python
from nltk.corpus import mac_morpho
mfc_news_tagged = mac_morpho.tagged_words()
mfc_tag_fd = nltk.FreqDist(tag for (word, tag) in mfc_news_tagged)
mfc_tag_fd.most_common()
```

**C) Na terceira célula deve ser gerado um gráfico não acumulativo das etiquetas mais comuns**.

```python
mfc_tag_fd.plot(cumulative=False)
```
